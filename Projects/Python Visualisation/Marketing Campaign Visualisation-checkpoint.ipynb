```python
import pandas as pd 
import matplotlib.pyplot as plt
import datetime

# 1. Зроби групування даних по дням та для згрупованих даних побудуй два графіки:
# Графік з щоденною сумою витрат на рекламу в 2021 році;
# Графік з щоденним ROMI в 2021 році.
adsData = pd.read_csv("facebook_ads_data.csv") 
adsData["ad_date"] = pd.to_datetime(adsData["ad_date"])

grouped1 = adsData[["total_spend", "total_value", "ad_date"]]
grouped1 = grouped1.groupby(by = "ad_date").agg({
    'total_value' : 'sum', 
    'total_spend' : 'sum'
})

plt.plot(grouped1["total_value"]/grouped1["total_spend"], linestyle = "-.", label = "romi")
plt.xlabel('Date', fontsize='small', color='midnightblue')
plt.ylabel('ROMI', fontsize='small', color='midnightblue')
plt.title("Daily ROMI")
plt.xticks(rotation=90)
plt.show()

plt.plot(
    grouped1["total_spend"],  
    linestyle = "-."
)
plt.xlabel('Date', fontsize='small', color='midnightblue')
plt.ylabel('Spend', fontsize='small', color='midnightblue')
plt.title("Daily spendings")
plt.xticks(rotation=90)
plt.show()
```

```python
import seaborn as sns
# 2. Зроби групування даних за назвою кампанії побудуй два графіки:
# Графік загальної суми витрат на рекламу в кожній з кампаній;
# Загальний ROMI в кожній з кампаній.

adsData = pd.read_csv("facebook_ads_data.csv") 
grouped2 = adsData[["campaign_name", "total_spend", "total_value"]]
grouped2 = grouped2.groupby(by = "campaign_name").agg({
    'total_spend' : 'sum', 
    'total_value' : 'sum'
}).reset_index()
grouped2["ROMI"] = grouped2["total_value"]/grouped2["total_spend"]
plt.bar(
    grouped2["campaign_name"],
    grouped2["ROMI"] 
)
plt.xlabel('Date', fontsize='small', color='midnightblue')
plt.ylabel('ROMI', fontsize='small', color='midnightblue')
plt.title("ROMI by Campaign")
plt.xticks(rotation=90)
plt.show()

plt.bar(
    grouped2["campaign_name"],
    grouped2["total_spend"]  
)
plt.xlabel('Date', fontsize='small', color='midnightblue')
plt.ylabel('Spend', fontsize='small', color='midnightblue')
plt.title("Spendings by Campaign")
plt.xticks(rotation=90)
plt.show()
```

```python
#3 За допомогою box plot визнач, яким був розкид щоденного ROMI в кожній кампаній (за назвою кампанії.)
adsData = pd.read_csv("facebook_ads_data.csv") 
sns.boxplot(data = adsData, x = "campaign_name", y = "romi")
plt.xticks(rotation=90)
```

```python
#5 Побудуй гістограму з розподілом значень ROMI в таблиці facebook_ads_data.csv
import seaborn as sns
import pandas as pd

adsData = pd.read_csv("facebook_ads_data.csv") 
histo_data = pd.DataFrame()
histo_data["romi"] = adsData["romi"].dropna()
sns.set_style("whitegrid")
sns.displot(data = histo_data, x = "romi")
plt.xlabel("romi")
plt.ylabel("frequency")
plt.title("histogram of romi distribution")
plt.show()
```

```python
# # Побудуй теплову карту кореляції між усіма числовими показниками в таблиці 
# facebook_ads_data.csv. Які показники мають найвищу та найнижчу кореляцію? З чим корелює "total_value"?
from IPython.display import display
corr = adsData.dropna()[["total_spend","total_impressions","total_clicks","total_value","cpc","cpm","ctr","romi"]].corr()
sns.heatmap(corr, annot = True, cmap = 'cividis')
print("Найсильніша кореляція показника total_value спостерігається з показником total_spend (98%)")
```

```python
#6 Побудуй точкову діаграму з лінійною регресією (для цього ти можеш використати функцію lmplot()), 
# що побудована на даних з "total_spend" та "total_value" щоб візуалізувати звʼязок між цими величинами.
sns.lmplot(data = adsData.dropna(), x = "total_spend", y = "total_value")
```
